```python
import torch
```

```python
t1 = torch.tensor(4.)

print(t1)
print("Data Type: ", t1.dtype)
print("Dimension: ", t1.ndim)
print("Shape: ", t1.shape)
```

# Vector

```python
t2 = torch.tensor([1, 2, 3, 4])

print(t2)
print("Data Type: ", t2.dtype)
print("Dimension: ", t2.ndim)
print("Shape: ", t2.shape)
```

# Matrix

```python
t3 = torch.tensor([[5., 6, 7],
                   [8, 9, 0],
                   [4, 7, 1]])

print(t3)
print("Data Type: ", t3.dtype)
print("Dimension: ", t3.ndim)
print("Shape: ", t3.shape)
```

# 3 Dim Array

```python
t4 = torch.tensor([[[15., 16, 17],
                   [18, 19, 10],
                   [14, 17, 11]],
                   [[15., 16, 17],
                   [18, 19, 10],
                   [14, 17, 11]]])

print(t4)
print("Data Type: ", t4.dtype)
print("Dimension: ", t4.ndim)
print("Shape: ", t4.shape)
```

# Tensor Operations & Gradients

```python
x = torch.tensor(3.)
w = torch.tensor(4., requires_grad=True)
b = torch.tensor(5., requires_grad=True)
```

## Basic Math Operation

```python
y = w * x + b
print(y)
```

## Compute Derivative

```python
y.backward()
```

```python
print("dy/dx: ", x.grad)
print("dy/dw: ", w.grad)
print("dy/db: ", b.grad)
```

# Pytorch & Numpy

```python
import numpy as np
```

```python
x = np.array([[1., 2.],
             [3., 4.]])
print(x)
```

## Convert numpy into tensor

```python
y = torch.from_numpy(x)
print(y)
```

```python
x.dtype, y.dtype
```

```python
y = torch.tensor(x)
print(y)
print(y.dtype)
```

## Convert pytorch tensor into numpy

```python
z = y.numpy()
print(z)
print(z.dtype)
```
